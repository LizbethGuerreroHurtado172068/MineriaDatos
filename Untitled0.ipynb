IPython Magic Commands

```python
 # Sirve para evaluar el desempeño de una expresión en cuanto a tiempo en una lista de comprensión
 %timeit L = [n ** 2 for n in range(1000)]
```

```python
#Con doble porcentaje podemos evaluar el mismo desmpeño en cuanto a tiempo en más de una linea de código
%%timeit
L = []
for n in range(1000):
  L.append(n ** 2)
```

Understanding Data Types in Python


```python
#Importa numpy y renombralo como np (la cual genera arreglos de cualquier dimensión)
import numpy as np
```

```python
#Arreglo lineal de enteros
np.array([1, 4, 2, 5, 3])
```

```python
#Al ingresar enteros y flotantes, los combierte en flotantes (segun el inicial de la lista)
np.array([3.14, 4, 2, 3])
```

```python
#Transforma el array en el tipo de dato que se le pida
np.array([1, 2, 3, 4], dtype='float32')
```

```python
np.array([1, 2, 3, 4], dtype='int')
```

```python
#Cuando el elemento de np.array es una iteración anidada genera una matriz
#Las listas anidadas dan como resultado matrices multidimensionales
np.array([range(i, i + 3) for i in [2, 4, 6]])
```
